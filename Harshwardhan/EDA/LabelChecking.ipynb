```python
import pandas as pd
import numpy as np


df_1=pd.read_excel('set3.xlsx')
df_2=pd.read_excel('set5.xlsx')
```

```python
columns_to_keep=['PW(µs)','Azimuth(º)','Elevation(º)','Power(dBm)','Freq(MHz)','Name']

df_1=df_1[columns_to_keep]
df_2=df_2[columns_to_keep]

```

```python
df_2['Name'].value_counts()
```

```python
df_3=pd.read_excel('set6.xlsx')
df_3=df_3[columns_to_keep]
df_3['Name'].value_counts()
```

```python
names_1 = set(df_1['Name'].unique())
names_2 = set(df_2['Name'].unique())
names_3 = set(df_3['Name'].unique())

common_names = names_1.intersection(names_2, names_3)

print(common_names)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import matplotlib.gridspec as gridspec

# Make plots look nicer
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("viridis")

# These are the columns we want to analyze
numerical_cols = ['PW(µs)', 'Azimuth(º)', 'Elevation(º)', 'Power(dBm)', 'Freq(MHz)']

for name in common_names:
    print(f"=== Analysis for {name} ===\n")
    
    # Filter dataframes for the current name
    df1_subset = df_1[df_1['Name'] == name]
    df2_subset = df_2[df_2['Name'] == name]
    df3_subset = df_3[df_3['Name'] == name]
    
    # Set up the figure for all visualizations
    fig = plt.figure(figsize=(20, len(numerical_cols)*4))
    gs = gridspec.GridSpec(len(numerical_cols), 3)
    
    # Analyze each column
    for i, col in enumerate(numerical_cols):
        print(f"\n--- {col} Statistics ---")
        
        # Extract the data for this column
        data1 = df1_subset[col]
        data2 = df2_subset[col]
        data3 = df3_subset[col]
        
        # Calculate statistics
        print("DataFrame 1:")
        print(data1.describe())
        print("\nDataFrame 2:")
        print(data2.describe())
        print("\nDataFrame 3:")
        print(data3.describe())
        
        # Calculate variability across datasets
        all_data = pd.concat([data1, data2, data3])
        overall_var = all_data.var()
        overall_range = all_data.max() - all_data.min()
        print(f"\nOverall variability: Variance = {overall_var:.4f}, Range = {overall_range:.4f}")
        
        # Create histograms
        ax1 = plt.subplot(gs[i, 0])
        ax1.set_title(f"{col} Distribution")
        sns.histplot(data1, color='blue', alpha=0.5, label='Set 1', ax=ax1)
        sns.histplot(data2, color='green', alpha=0.5, label='Set 2', ax=ax1)
        sns.histplot(data3, color='red', alpha=0.5, label='Set 3', ax=ax1)
        ax1.legend()
        
        # Create box plots
        ax2 = plt.subplot(gs[i, 1])
        ax2.set_title(f"{col} Box Plot")
        data_to_plot = [data1, data2, data3]
        bp = ax2.boxplot(data_to_plot, patch_artist=True, labels=['Set 1', 'Set 2', 'Set 3'])
        
        # Color the boxplots
        colors = ['blue', 'green', 'red']
        for patch, color in zip(bp['boxes'], colors):
            patch.set_facecolor(color)
            patch.set_alpha(0.5)
        
        # Create violin plots
        ax3 = plt.subplot(gs[i, 2])
        ax3.set_title(f"{col} Violin Plot")
        df_combined = pd.DataFrame({
            'Value': pd.concat([data1, data2, data3]),
            'Dataset': ['Set 1']*len(data1) + ['Set 2']*len(data2) + ['Set 3']*len(data3)
        })
        sns.violinplot(x='Dataset', y='Value', data=df_combined, palette=['blue', 'green', 'red'], ax=ax3)
        
        # Add mean markers to violin plots
        means = [data1.mean(), data2.mean(), data3.mean()]
        ax3.scatter([0, 1, 2], means, marker='o', color='black', s=30, zorder=3)
        
    plt.tight_layout()
    plt.suptitle(f"Feature Analysis for {name}", fontsize=16, y=1.02)
    plt.show()
    
    # Create correlation heatmaps for each dataset
    plt.figure(figsize=(20, 6))
    
    plt.subplot(1, 3, 1)
    sns.heatmap(df1_subset[numerical_cols].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title(f'Correlations in Set 1 - {name}')
    
    plt.subplot(1, 3, 2)
    sns.heatmap(df2_subset[numerical_cols].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title(f'Correlations in Set 2 - {name}')
    
    plt.subplot(1, 3, 3)
    sns.heatmap(df3_subset[numerical_cols].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title(f'Correlations in Set 3 - {name}')
    
    plt.tight_layout()
    plt.show()
    
    # Create scatter plot matrix for the first few data points to see relationships between features
    # Limit to prevent overplotting
    limit = min(100, min(len(df1_subset), len(df2_subset), len(df3_subset)))
    
    combined_df = pd.DataFrame({
        **{f"{col}_Set1": df1_subset[col].iloc[:limit].values for col in numerical_cols},
        **{f"{col}_Set2": df2_subset[col].iloc[:limit].values for col in numerical_cols},
        **{f"{col}_Set3": df3_subset[col].iloc[:limit].values for col in numerical_cols},
    })
    
    # Focus on one important feature pair (freq and power) across datasets
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.scatter(df1_subset['Freq(MHz)'], df1_subset['Power(dBm)'], alpha=0.5)
    plt.title(f'Set 1: Freq vs Power - {name}')
    plt.xlabel('Freq(MHz)')
    plt.ylabel('Power(dBm)')
    
    plt.subplot(1, 3, 2)
    plt.scatter(df2_subset['Freq(MHz)'], df2_subset['Power(dBm)'], alpha=0.5, color='green')
    plt.title(f'Set 2: Freq vs Power - {name}')
    plt.xlabel('Freq(MHz)')
    plt.ylabel('Power(dBm)')
    
    plt.subplot(1, 3, 3)
    plt.scatter(df3_subset['Freq(MHz)'], df3_subset['Power(dBm)'], alpha=0.5, color='red')
    plt.title(f'Set 3: Freq vs Power - {name}')
    plt.xlabel('Freq(MHz)')
    plt.ylabel('Power(dBm)')
    
    plt.tight_layout()
    plt.show()

    print("\n" + "="*50 + "\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import combinations
import math

for name in common_names:
    print(f"=== Feature Pair Analysis for {name} ===\n")
    
    # Filter dataframes for the current name
    df1_subset = df_1[df_1['Name'] == name]
    df2_subset = df_2[df_2['Name'] == name]
    df3_subset = df_3[df_3['Name'] == name]
    
    # Get all unique pairs of features
    feature_pairs = list(combinations(numerical_cols, 2))
    
    # Calculate number of rows and columns for the subplot grid
    n_pairs = len(feature_pairs)
    n_cols = 3  # One column for each dataset
    n_rows = math.ceil(n_pairs / 1)  # One pair per row
    
    # Create scatter plots for all feature pairs across datasets
    for idx, (feature1, feature2) in enumerate(feature_pairs):
        fig, axes = plt.subplots(1, 3, figsize=(20, 6))
        fig.suptitle(f"{feature1} vs {feature2} - {name}", fontsize=16)
        
        # Set 1
        axes[0].scatter(df1_subset[feature1], df1_subset[feature2], alpha=0.5, color='blue')
        axes[0].set_title(f'Set 1')
        axes[0].set_xlabel(feature1)
        axes[0].set_ylabel(feature2)
        
        # Set 2
        axes[1].scatter(df2_subset[feature1], df2_subset[feature2], alpha=0.5, color='green')
        axes[1].set_title(f'Set 2')
        axes[1].set_xlabel(feature1)
        axes[1].set_ylabel(feature2)
        
        # Set 3
        axes[2].scatter(df3_subset[feature1], df3_subset[feature2], alpha=0.5, color='red')
        axes[2].set_title(f'Set 3')
        axes[2].set_xlabel(feature1)
        axes[2].set_ylabel(feature2)
        
        plt.tight_layout()
        plt.show()
    
    # Create a radar chart to compare feature variations across datasets
    print("\nRadar Chart Analysis of Feature Variations\n")
    
    # Calculate coefficient of variation (CV) for each feature in each dataset
    cv_data = {}
    for col in numerical_cols:
        cv_data[col] = {
            'Set 1': df1_subset[col].std() / abs(df1_subset[col].mean()) if abs(df1_subset[col].mean()) > 0 else 0,
            'Set 2': df2_subset[col].std() / abs(df2_subset[col].mean()) if abs(df2_subset[col].mean()) > 0 else 0,
            'Set 3': df3_subset[col].std() / abs(df3_subset[col].mean()) if abs(df3_subset[col].mean()) > 0 else 0
        }
    
    # Create radar chart
    categories = numerical_cols
    N = len(categories)
    
    # Create angles for each feature
    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += angles[:1]  # Close the loop
    
    # Create the plot
    fig = plt.figure(figsize=(10, 10))
    ax = plt.subplot(111, polar=True)
    
    # Draw one axis per variable and add labels
    plt.xticks(angles[:-1], categories, size=12)
    
    # Draw the coefficient of variation for each dataset
    for dataset_idx, dataset_name in enumerate(['Set 1', 'Set 2', 'Set 3']):
        values = [cv_data[col][dataset_name] for col in categories]
        values += values[:1]  # Close the loop
        
        # Plot values
        ax.plot(angles, values, linewidth=2, linestyle='solid', label=dataset_name)
        ax.fill(angles, values, alpha=0.1)
    
    # Add legend
    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
    plt.title(f"Feature Variation Comparison for {name}", size=15)
    plt.show()
    
    # Create 3D scatter plots for selected feature triplets
    if len(numerical_cols) >= 3:
        print("\n3D Visualization of Selected Features\n")
        
        # Choose a representative triplet of features
        triplet = [numerical_cols[0], numerical_cols[1], numerical_cols[2]]
        
        fig = plt.figure(figsize=(20, 7))
        
        # Set 1
        ax1 = fig.add_subplot(131, projection='3d')
        ax1.scatter(df1_subset[triplet[0]], df1_subset[triplet[1]], df1_subset[triplet[2]], 
                   alpha=0.6, c='blue')
        ax1.set_xlabel(triplet[0])
        ax1.set_ylabel(triplet[1])
        ax1.set_zlabel(triplet[2])
        ax1.set_title(f'Set 1 - {name}')
        
        # Set 2
        ax2 = fig.add_subplot(132, projection='3d')
        ax2.scatter(df2_subset[triplet[0]], df2_subset[triplet[1]], df2_subset[triplet[2]], 
                   alpha=0.6, c='green')
        ax2.set_xlabel(triplet[0])
        ax2.set_ylabel(triplet[1])
        ax2.set_zlabel(triplet[2])
        ax2.set_title(f'Set 2 - {name}')
        
        # Set 3
        ax3 = fig.add_subplot(133, projection='3d')
        ax3.scatter(df3_subset[triplet[0]], df3_subset[triplet[1]], df3_subset[triplet[2]], 
                   alpha=0.6, c='red')
        ax3.set_xlabel(triplet[0])
        ax3.set_ylabel(triplet[1])
        ax3.set_zlabel(triplet[2])
        ax3.set_title(f'Set 3 - {name}')
        
        plt.tight_layout()
        plt.show()
    
    # Create parallel coordinates plot to visualize all features simultaneously
    print("\nParallel Coordinates Plot\n")
    
    # Prepare data for parallel coordinates
    df1_sample = df1_subset.sample(min(100, len(df1_subset))).copy()
    df1_sample['Dataset'] = 'Set 1'
    
    df2_sample = df2_subset.sample(min(100, len(df2_subset))).copy()
    df2_sample['Dataset'] = 'Set 2'
    
    df3_sample = df3_subset.sample(min(100, len(df3_subset))).copy()
    df3_sample['Dataset'] = 'Set 3'
    
    # Combine the data
    combined_df = pd.concat([df1_sample, df2_sample, df3_sample])
    
    # Create parallel coordinates plot
    plt.figure(figsize=(15, 10))
    pd.plotting.parallel_coordinates(combined_df, 'Dataset', cols=numerical_cols, 
                                    color=['blue', 'green', 'red'], alpha=0.3)
    plt.title(f'Parallel Coordinates Plot for {name}')
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
    
    print("\n" + "="*50 + "\n")
```
